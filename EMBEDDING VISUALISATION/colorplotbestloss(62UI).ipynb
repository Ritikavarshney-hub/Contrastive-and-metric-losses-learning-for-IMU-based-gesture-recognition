```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, ConcatDataset
from torch.optim import Adam

from pytorch_metric_learning.losses import SupConLoss
from pytorch_metric_learning.miners import BatchEasyHardMiner

import umap
import matplotlib.pyplot as plt
from collections import defaultdict

torch.manual_seed(42)
np.random.seed(42)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

IDX2CHAR = {
    0: "A", 1: "B", 2: "C/c", 3: "D", 4: "E", 5: "F", 6: "G",
    7: "H", 8: "I", 9: "J", 10: "K/k", 11: "L", 12: "M/m",
    13: "N/n", 14: "O/o/0", 15: "P/p", 16: "Q", 17: "R",
    18: "S/s", 19: "T", 20: "U/u", 21: "V/v", 22: "W/w",
    23: "X/x", 24: "Y/y", 25: "Z/z",
    26: "a", 27: "b", 28: "d", 29: "e", 30: "f",
    31: "g/9", 32: "h", 33: "i", 34: "j", 35: "l",
    36: "q", 37: "r", 38: "t",
    39: "1", 40: "2", 41: "3", 42: "4",
    43: "5", 44: "6", 45: "7", 46: "8"
}

class EMGDataset(Dataset):
    def __init__(self, x_path, y_path):
        X = np.load(x_path)
        y = np.argmax(np.load(y_path), axis=1)

        for i in range(X.shape[0]):
            for c in range(6):
                X[i, :, c] = (X[i, :, c] - X[i, :, c].mean()) / (X[i, :, c].std() + 1e-8)

        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.long)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

class CNNEncoder(nn.Module):
    def __init__(self, emb_dim=256):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(6, 64, 10), nn.ReLU(),
            nn.Conv1d(64, 64, 10), nn.ReLU(),
            nn.MaxPool1d(3),

            nn.Conv1d(64, 256, 10), nn.ReLU(),
            nn.Conv1d(256, 256, 10), nn.ReLU(),

            nn.AdaptiveAvgPool1d(1)
        )
        self.fc = nn.Linear(256, emb_dim)

    def forward(self, x):
        x = x.permute(0, 2, 1)
        x = self.net(x).squeeze(-1)
        return self.fc(x)

class FullModel(nn.Module):
    def __init__(self, emb_dim=256, num_classes=47):
        super().__init__()
        self.encoder = CNNEncoder(emb_dim)
        self.classifier = nn.Linear(emb_dim, num_classes)

    def forward(self, x):
        z = self.encoder(x)
        logits = self.classifier(z)
        return z, logits

def train_scl(model, loader, epochs=50):
    supcon = SupConLoss(temperature=0.1)
    miner = BatchEasyHardMiner(pos_strategy="semihard", neg_strategy="hard")
    ce = nn.CrossEntropyLoss()
    opt = Adam(model.parameters(), lr=5e-4)

    for epoch in range(epochs):
        model.train()
        total_loss, correct, total = 0, 0, 0

        for x, y in loader:
            x, y = x.to(device), y.to(device)

            z, logits = model(x)
            z = F.normalize(z, dim=1)

            idx = miner(z, y)
            loss = supcon(z, y, idx) + ce(logits, y)

            opt.zero_grad()
            loss.backward()
            opt.step()

            total_loss += loss.item()
            correct += (logits.argmax(1) == y).sum().item()
            total += y.size(0)

        print(
            f"Epoch {epoch+1:02d} | "
            f"Loss: {total_loss/len(loader):.4f} | "
            f"Acc: {correct/total:.4f}"
        )

def extract_embeddings(model, loader):
    model.eval()
    Z, Y = [], []

    with torch.no_grad():
        for x, y in loader:
            x = x.to(device)
            z, _ = model(x)
            z = F.normalize(z, dim=1)
            Z.append(z.cpu().numpy())
            Y.append(y.numpy())

    Z = np.vstack(Z)
    Y = np.hstack(Y)

    np.save("embeddings_scl_47.npy", Z)
    np.save("labels_scl_47.npy", Y)

    print("Saved embeddings_scl_47.npy and labels_scl_47.npy")
    print("Embedding shape:", Z.shape)

    return Z, Y

def plot_umap(Z, Y, title):
    reducer = umap.UMAP(
        n_components=2,
        n_neighbors=40,
        min_dist=0.25,
        spread=1.5,
        metric="cosine",
        random_state=42
    )

    Z_2d = reducer.fit_transform(Z)

    plt.figure(figsize=(10, 10))
    plt.scatter(
        Z_2d[:, 0], Z_2d[:, 1],
        c=Y, cmap="tab20",
        s=8, alpha=0.6, linewidths=0
    )

    centers = defaultdict(list)
    for i, label in enumerate(Y):
        centers[int(label)].append(Z_2d[i])

    for label, pts in centers.items():
        pts = np.array(pts)
        cx, cy = pts.mean(axis=0)
        plt.text(
            cx, cy, IDX2CHAR[label],
            fontsize=10, weight="bold",
            ha="center", va="center",
            bbox=dict(facecolor="white", alpha=0.75, edgecolor="none")
        )

    plt.title(title, fontsize=14)
    plt.axis("off")
    plt.tight_layout()
    plt.show()
    
BASE = "models/Data/Data/47_classes/UserDependenet"

train_ds = EMGDataset(
    f"{BASE}/Train/X_train_1.npy",
    f"{BASE}/Train/y_train_1.npy"
)
test_ds = EMGDataset(
    f"{BASE}/Test/X_test_1.npy",
    f"{BASE}/Test/y_test_1.npy"
)

merged_ds = ConcatDataset([train_ds, test_ds])
merged_loader = DataLoader(merged_ds, batch_size=128, shuffle=True)

model = FullModel().to(device)
print(model)
train_scl(model, merged_loader, epochs=50)

embed_loader = DataLoader(merged_ds, batch_size=128, shuffle=False)
Z, Y = extract_embeddings(model, embed_loader)

plot_umap(
    Z, Y,
    title="UMAP of CNN Embeddings (SupCon, 47 Merged Classes),User Dependent"
)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------- LOAD ONE SAMPLE ----------------
BASE = "models/Data/Data/62_classes/UserDependenet"

X = np.load(f"{BASE}/Train/X_train_1.npy")   # shape: (N, T, 6)

sample_idx = 0        # choose any index
raw_emg = X[sample_idx]   # (T, 6)

# ---------------- NORMALIZATION ----------------
norm_emg = np.zeros_like(raw_emg)

for c in range(6):
    mu = raw_emg[:, c].mean()
    std = raw_emg[:, c].std()
    norm_emg[:, c] = (raw_emg[:, c] - mu) / (std + 1e-8)

# ---------------- PLOTTING ----------------
channels = 6
time = np.arange(raw_emg.shape[0])

fig, axes = plt.subplots(channels, 2, figsize=(14, 10), sharex=True)

for c in range(channels):
    # Raw IMU
    axes[c, 0].plot(time, raw_emg[:, c], color="tab:blue")
    axes[c, 0].set_title(f"Raw IMU - Channel {c+1}")
    axes[c, 0].set_ylabel("Amplitude")

    # Normalized IMU
    axes[c, 1].plot(time, norm_emg[:, c], color="tab:orange")
    axes[c, 1].set_title(f"Normalized IMU - Channel {c+1}")

axes[-1, 0].set_xlabel("Time (samples)")
axes[-1, 1].set_xlabel("Time (samples)")

plt.suptitle("Raw vs Normalized IMU Signals (Single Sample)", fontsize=14)
plt.tight_layout()
plt.show()
```
