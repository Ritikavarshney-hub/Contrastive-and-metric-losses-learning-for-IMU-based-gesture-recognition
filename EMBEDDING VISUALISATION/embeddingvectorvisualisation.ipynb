```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, ConcatDataset
from torch.optim import Adam

import matplotlib.pyplot as plt
import umap
from collections import defaultdict

from pytorch_metric_learning import losses, miners, distances

torch.manual_seed(42)
np.random.seed(42)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

class EMGDataset(Dataset):
    def __init__(self, x_path, y_path):
        X = np.load(x_path)
        y = np.load(y_path)

        for i in range(X.shape[0]):
            for c in range(6):
                X[i, :, c] = (X[i, :, c] - X[i, :, c].mean()) / (X[i, :, c].std() + 1e-8)

        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(np.argmax(y, axis=1), dtype=torch.long)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

class CNNEncoder(nn.Module):
    def __init__(self, emb_dim=256):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(6, 64, kernel_size=10),
            nn.ReLU(),
            nn.Conv1d(64, 64, kernel_size=10),
            nn.ReLU(),
            nn.MaxPool1d(3),

            nn.Conv1d(64, 256, kernel_size=10),
            nn.ReLU(),
            nn.Conv1d(256, 256, kernel_size=10),
            nn.ReLU(),

            nn.AdaptiveAvgPool1d(1)
        )
        self.fc = nn.Linear(256, emb_dim)

    def forward(self, x):
        x = x.permute(0, 2, 1)          # (B, C, T)
        x = self.net(x).squeeze(-1)     # (B, 256)
        z = self.fc(x)
        return z

class FullModel(nn.Module):
    def __init__(self, emb_dim=256, num_classes=62):
        super().__init__()
        self.encoder = CNNEncoder(emb_dim)
        self.classifier = nn.Linear(emb_dim, num_classes)

    def forward(self, x):
        z = self.encoder(x)
        logits = self.classifier(z)
        return z, logits

def train(model, loader, metric_loss, miner, epochs=40):
    optimizer = Adam(model.parameters(), lr=5e-4)
    ce = nn.CrossEntropyLoss()

    model.train()
    for epoch in range(epochs):
        for x,y in loader:
            x,y = x.to(device), y.to(device)
            z, logits = model(x)
            z = F.normalize(z, dim=1)

            idx = miner(z, y)
            loss = metric_loss(z, y, idx) + ce(logits, y)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        print(f"Epoch {epoch+1}/{epochs} | Loss {loss.item():.4f}")

def extract_embeddings(model, loader):
    model.eval()
    Z, Y = [], []

    with torch.no_grad():
        for x,y in loader:
            z,_ = model(x.to(device))
            Z.append(F.normalize(z,dim=1).cpu().numpy())
            Y.append(y.numpy())

    return np.vstack(Z), np.concatenate(Y)

def plot_umap(Z, Y, title):
    reducer = umap.UMAP(
        n_neighbors=30,
        min_dist=0.05,
        n_components=2,
        random_state=42
    )
    Z2 = reducer.fit_transform(Z)

    plt.figure(figsize=(9,9))
    plt.scatter(Z2[:,0], Z2[:,1], c=Y, cmap="tab20", s=6)

    centers = defaultdict(list)
    for p,y in zip(Z2,Y):
        centers[y].append(p)

    for cls, pts in centers.items():
        pts = np.array(pts)
        cx, cy = pts.mean(axis=0)
        plt.text(cx, cy, str(cls), fontsize=10, weight="bold")

    plt.title(title)
    plt.axis("off")
    plt.show()

BASE = "models/Data/Data/62_classes/UserIndependent"

train_ds = EMGDataset(f"{BASE}/Train/X_train_1.npy",
                      f"{BASE}/Train/y_train_1.npy")
test_ds  = EMGDataset(f"{BASE}/Test/X_test_1.npy",
                      f"{BASE}/Test/y_test_1.npy")

full_ds = ConcatDataset([train_ds, test_ds])
loader = DataLoader(full_ds, batch_size=128, shuffle=True)

losses_dict = {
    "Triplet_L2": losses.TripletMarginLoss(
        margin=0.2,
        distance=distances.LpDistance(normalize_embeddings=True)
    ),
    "Triplet_Cosine": losses.TripletMarginLoss(
        margin=0.1,
        distance=distances.CosineSimilarity()
    ),
    "NTXent": losses.NTXentLoss(temperature=0.2),
    "SupCon": losses.SupConLoss(temperature=0.1)
}

miner = miners.BatchEasyHardMiner(
    pos_strategy="semihard",
    neg_strategy="hard"
)

for loss_name, metric_loss in losses_dict.items():
    print(f"\n===== {loss_name} =====")

    model = FullModel().to(device)
    train(model, loader, metric_loss, miner)

    Z, Y = extract_embeddings(model, loader)
    plot_umap(Z, Y, f"{loss_name}  UMAP (62 classes)")

    del model
    torch.cuda.empty_cache()


```
